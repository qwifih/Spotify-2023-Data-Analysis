```python
#import the functions needed for this program
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Overview of Dataset

```python
#reads the csv file with the data necessary for this program
dataFrame = pd.read_csv('C:\\Users\\USER\Documents\\Python Scripts\\spotify-2023.csv', encoding='latin-1') #reads the csv file of the dataset of the most streamed spotify songs in 2023
df = pd.DataFrame(dataFrame,columns=['track_name','artist(s)_name','artist_count','released_year','released_month','released_day','in_spotify_playlists','in_spotify_charts','streams','in_apple_playlists','in_apple_charts','in_deezer_playlists','in_deezer_charts,in_shazam_charts','bpm','key','mode','danceability_%','valence_%','energy_%','acousticness_%','instrumentalness_%','liveness_%','speechiness_%'])
df
```

## Basic Descriptive Statistics

```python
#fixes the eror appearing
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

#basic descriptive statistics for the streams column
df['streams'] = pd.to_numeric(df['streams'])

#calculate mean, median, and standard deviation of the streams column
mean = df['streams'].mean()
median = df['streams'].median()
STD = df['streams'].std()

#shows the results of the calculated mean, median, mode
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {STD}")
```

```python
#basic descriptive statistics of the years between 2015 and 2024
df_filtered = df[(df['released_year'] >= 2015) & (df['released_year'] <= 2024)]
df_filtered
```

```python
#sets up figure for the subplots
plt.figure() 

#histogram for released_year
plt.subplot()
sns.histplot(df_filtered['released_year'], bins=25, kde=True)
plt.title('Distribution of Released Year (2015-2024)')
plt.xlabel('Released Year')
plt.ylabel('Frequency')
plt.xlim(2015, 2024)  

#boxplot for artist count
plt.figure(figsize=(10, 7))
sns.boxplot(x=df['artist_count'], color='#DAF0F7', fliersize=5)
plt.title('Distribution of Artist Count', fontsize=18, fontweight='bold')
plt.xlabel('Artist Count', fontsize=14)
plt.grid(axis = 'x')

#display the graph
plt.show()
```

## Temporal Trends

```python
#filter the released years to get data from 2000 to 2023
df_filtered = df[(df['released_year'] >= 2000) & (df['released_year'] <= 2023)]

#counts the tracks per year
tpy = df_filtered['released_year'].value_counts().sort_index()

#create a line graph for the number of tracks released per year
plt.figure(figsize=(12, 6))
plt.plot(tpy.index, tpy.values, marker='o', linestyle='-')
plt.title('Number of Tracks Released Over Time (2000-2023)')
plt.xlabel('Year')
plt.ylabel('Number of Tracks Released')
plt.xticks(tpy.index, rotation=45)
plt.grid()

#display the graph
plt.show
```

```python
#the trends for tracks per month
tpm = df['released_month'].value_counts().sort_index()

#create a bar graph for the number of tracks released per month
plt.figure(figsize=(12, 6))
plt.bar(tpm.index, tpm.values, color='#DAF0F7')
plt.title('Number of Tracks Released Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Tracks Released')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                       'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y')

#display the graph
plt.show()

#identifies the month with the most releases
most_releases_month = tpm.idxmax()
most_releases_count = tpm.max()
print(f"The month with the most releases is: {most_releases_month} with {most_releases_count} releases.")
```

## Genre and Music Characteristics

```python
#converts the relevant columns to numberic
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
df['bpm'] = pd.to_numeric(df['bpm'], errors='coerce')
df['danceability_%'] = pd.to_numeric(df['danceability_%'], errors='coerce')
df['energy_%'] = pd.to_numeric(df['energy_%'], errors='coerce')

#check for conversion
print("Number of NaN values in streams column after conversion:", df['streams'].isna().sum())
print("Number of NaN values in bpm column after conversion:", df['bpm'].isna().sum())
print("Number of NaN values in danceability_% column after conversion:", df['danceability_%'].isna().sum())
print("Number of NaN values in energy_% column after conversion:", df['energy_%'].isna().sum())


#choose the necessary columns for correlation analysis
correlation_data = df[['streams', 'bpm', 'danceability_%', 'energy_%']].dropna()

#calculates the correlation matrix
correlation_mat = correlation_data.corr()

#create a heatmap to better visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_mat, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation between Streams and Musical Attributes')
plt.show()

#identifies the attributes with the highest correlation with streams and value
correlation_with_streams = correlation_mat['streams'].drop('streams')
highest_correlation = correlation_with_streams.abs().idxmax()
highest_correlation_value = correlation_with_streams[highest_correlation]

print(f"The attribute that influences streams the most is: '{highest_correlation}' with a correlation of {highest_correlation_value:.2f}.")
```

```python
#to make the graph clearer
plt.figure(figsize=(12, 5))

#create a scatter plot for Danceability vs Energy
plt.subplot(1, 2, 1)
sns.scatterplot(x='danceability_%', y='energy_%', data=df, color='purple')
plt.title('Danceability (%) vs Energy (%)')
plt.xlabel('Danceability (%)')
plt.ylabel('Energy (%)')

#create a scatter plot for Valence vs Acousticness
plt.subplot(1, 2, 2)
sns.scatterplot(x='valence_%', y='acousticness_%', data=df, color='green')
plt.title('Valence (%) vs Acousticness (%)')
plt.xlabel('Valence (%)')
plt.ylabel('Acousticness (%)')

#display the graph
plt.show()
```

## Platform Popularity

```python
numeric_columns = ['in_spotify_playlists', 'in_spotify_charts', 'in_apple_playlists', 'streams']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col])

#fill in the NaN values for binary columns
df['in_spotify_playlists'] = df['in_spotify_playlists'].fillna(0)
df['in_spotify_charts'] = df['in_spotify_charts'].fillna(0)
df['in_apple_playlists'] = df['in_apple_playlists'].fillna(0)

#calculates the number of tracks in each category
track_counts = {
    'Spotify Playlists': df['in_spotify_playlists'].sum(),
    'Spotify Charts': df['in_spotify_charts'].sum(),
    'Apple Playlists': df['in_apple_playlists'].sum()
}

#create a bar plot to compare the number of tracks
plt.figure(figsize=(14, 6))
plt.bar(track_counts.keys(), track_counts.values(), color=['#DAF0F7', '#FFFF00', '#FFDFDD'])
plt.title('Number of Tracks in Different Platforms')
plt.ylabel('Number of Tracks')
plt.show()

#calculate average streams for tracks in each category
avg_streams = {
    'Spotify Playlists': df[df['in_spotify_playlists'] > 0]['streams'].mean(),
    'Spotify Charts': df[df['in_spotify_charts'] > 0]['streams'].mean(),
    'Apple Playlists': df[df['in_apple_playlists'] > 0]['streams'].mean()
}

#shows the average streams
print("Average Streams for Each Platform:")
for platform, avg in avg_streams.items():
    print(f"{platform}: {avg:.2f} streams")
```

## Advanced Analysis

```python
df['streams'] = pd.to_numeric(df['streams'],)

#group by key and mode, and calculate average streams
key_mode_streams = df.groupby(['key', 'mode'])['streams'].agg(['mean', 'median', 'std', 'count']).reset_index()

#create a bar plot for average streams by key and mode
plt.figure()
sns.barplot(data=key_mode_streams, x='key', y='mean', hue='mode', palette='muted')
plt.title('Average Streams by Key and Mode')
plt.xlabel('Key')
plt.ylabel('Average Streams')
plt.xticks(rotation=45)
plt.legend(title='Mode')
plt.grid(axis='y')

#display the graph
plt.show()
```

```python
artist_playlists = df.groupby('artist(s)_name')['in_spotify_playlists'].sum().reset_index()
artist_charts = df.groupby('artist(s)_name')['in_spotify_charts'].sum().reset_index()

#rename the columns for clarity
artist_playlists.rename(columns={'in_spotify_playlists': 'playlist_count'}, inplace=True)
artist_charts.rename(columns={'in_spotify_charts': 'chart_count'}, inplace=True)

#merge the two dataframes on artist name
artist_summary = pd.merge(artist_playlists, artist_charts, on='artist(s)_name', how='outer').fillna(0)
artist_summary['playlist_count'] = artist_summary['playlist_count'].astype(int)
artist_summary['chart_count'] = artist_summary['chart_count'].astype(int)

#sort by playlist and chart counts
artist_summary = artist_summary.sort_values(by=['playlist_count', 'chart_count'], ascending=False)

#display the top 10 artists
artist_summary.head(10) 
```
